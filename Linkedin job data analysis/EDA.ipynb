Load and Analyze Data in Pandas

```python
import pandas as pd

# Load CSV file into a DataFrame
df = pd.read_csv("linkedin_jobs.csv")

# Display first 5 rows
print(df.head())
```

```python
location_counts = df["Location"].value_counts()
print(location_counts)
```

```python
company_counts = df["Company"].value_counts()
print(company_counts.head(10))  # Show top 10 companies
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 job locations
top_locations = df["Location"].value_counts().head(10)

# Create a bar chart
plt.figure(figsize=(10,5))
sns.barplot(x=top_locations.index, y=top_locations.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Location")
plt.ylabel("Job Count")
plt.title("Top 10 Job Locations")
plt.show()
```

```python
top_companies = df["Company"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_companies.index, y=top_companies.values, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("Company")
plt.ylabel("Job Count")
plt.title("Top 10 Hiring Companies")
plt.show()
```
